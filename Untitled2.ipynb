```python
%matplotlib inline

import numpy as np
import base64
import matplotlib.pyplot as plt
from Crypto.Cipher import ARC4
from Crypto import Random
```

```python
# ****** UNKNOWN PARAMETERS
secret_cookie_b64 = 'QkUgU1VSRSBUTyBEUklOSyBZT1VSIE9WQUxUSU5F'
secret_cookie = base64.b64decode(secret_cookie_b64)
# ****** UNKNOWN PARAMETERS

def Challenge56_Oracle(request):
    
    key = Random.get_random_bytes(16)
    cipher = ARC4.new(key)
    return(cipher.encrypt(request + secret_cookie))
    
```

```python
score_mat = np.zeros((48, 256))
for ii in range(2, 48):
    score_mat[ii][0] = 1
    score_mat[ii][ii+1] = 1

score_mat[2][0] = 2
score_mat[2][130] = 1
score_mat[15][240] = 2
score_mat[31][224] = 2
score_mat[47][208] = 2
```

```python
Num_Iterations = 2**20

# See what length is if we provide a null "request"
secret_length = len(Challenge56_Oracle(b''))
pt_scores = np.zeros((secret_length, 256))

for pad_length in range(0, 48):
    
    print(f'{pad_length}, ', end='')
    request = b'\x00'*pad_length
    
    for jj in range(Num_Iterations):
        
        ciphertext = Challenge56_Oracle(request)
        
        for idx in range(pad_length+1, len(ciphertext)-pad_length):            
            CT_byte = ciphertext[idx]
            for PT_byte_guess in range(0, 256):
                key_guess = PT_byte_guess ^ CT_byte
                pt_scores[idx][] += score_mat[idx][key_guess]
                
```
